```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(6)
y = x ** 2
# creating fig and axes using subplots()
fig, axes = plt.subplots(figsize=(15,10))

# adding plots on axes
axes.plot(x, x**2, x, x**3, x, x*2, lw=2)

# adding x-ticks and x-tick-labels with some useful parameters
axes.set_xticks([1, 2, 3, 4, 5])
axes.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'],
                     fontsize=18)

# adding y-ticks and y-tick-labels
yticks = [0, 50, 100, 150]
axes.set_yticks(yticks)
# notice the list comprehension below inside "set_yticklabels"!
axes.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18);
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(5,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log") # setting our y scale to log scale, its super easy!
axes[1].set_title("Logarithmic scale (y)");
plt.show()
```

```python
fig, axes = plt.subplots()
      
axes.plot(x, x**2, x, np.exp(x))
axes.set_title("scientific notation")

axes.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
#  the ticker module supports completely configurable tick locating and formatting.
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) # setting scientific True
formatter.set_powerlimits((-2,2)) 
axes.yaxis.set_major_formatter(formatter) 
# To do: Change the above parameters and see how the plot changes, it's a good way to learn!
```

```python
fig, axes = plt.subplots(1, 1)
      
axes.plot(x, x**2, x, np.exp(x))
axes.set_yticks([0, 50, 100, 150])

axes.set_title("Graph_Title")
axes.set_xlabel("x")
axes.set_ylabel("y")

fig.subplots_adjust(left=0.15, right=0.9, bottom=0.1, top=0.9)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# customized grid
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='r', linestyle='--', linewidth=1, alpha=0.5)
plt.show()
```

```python
fig, axes1 = plt.subplots()

axes1.plot(x, x**2, lw=2, color="blue")
axes1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in axes1.get_yticklabels():
    label.set_color("blue")
    
axes2 = axes1.twinx()
axes2.plot(x, x**3, lw=2, color="red")
axes2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in axes2.get_yticklabels():
    label.set_color("red")
plt.show()
```

```python
fig, axes = plt.subplots()

axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')

axes.xaxis.set_ticks_position('bottom')

# setting position of x spine to x=0
axes.spines['bottom'].set_position(('data',0)) 

axes.yaxis.set_ticks_position('left')

# setting position of y spine to y=0
axes.spines['left'].set_position(('data',0))   

xx = np.linspace(-100.75, 100., 100)
axes.plot(xx, xx**3)
plt.show()
```

```python
fig = plt.figure()
axes1 = plt.subplot2grid((3,3), (0,0), colspan=3)
axes2 = plt.subplot2grid((3,3), (1,0), colspan=2)
axes3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
axes4 = plt.subplot2grid((3,3), (2,0))
axes5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
plt.show()
```

```python
# gridspec
import matplotlib.gridspec as gridspec
fig = plt.figure()

gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
plt.show()
```
